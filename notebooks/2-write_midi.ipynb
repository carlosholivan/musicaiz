# Writing a MIDI file from scratch

```python
import sys
sys.path.insert(1, "../")

from musicaiz.structure import Note, Instrument, Bar
from musicaiz.rhythm import TimeSignature
from musicaiz.loaders import Musa
```

```python
musa_obj = Musa()

musa_obj.time_sig = TimeSignature("4/4")

musa_obj.instruments.append(
    Instrument(
        program=30,
        general_midi=True
    )
)
n_bars = 1
for _ in range(0, n_bars, 1):
    musa_obj.instruments[0].bars.append(Bar())
        
notes_bar1 = [
    Note(pitch=69, start=0.5, end=1.0, velocity=127),
    Note(pitch=64, start=0.5, end=1.5, velocity=127),
    Note(pitch=67, start=1.0, end=1.5, velocity=127),
    Note(pitch=64, start=1.5, end=2.0, velocity=127)
]
musa_obj.instruments[0].bars[0].notes = notes_bar1
```

```python
musa_obj.instruments[0].bars[0].notes
```

```python
from musicaiz.rhythm import get_subdivisions

get_subdivisions(
    total_bars=1,
    subdivision="eight"
)
```

```python
musa_obj.write_midi("example")
```

```python
from musicaiz.plotters import Pianoroll
plot = Pianoroll()
plot.plot_instrument(
    total_bars=1,
    track=musa_obj.instruments[0].bars[0].notes,
    subdivision="eight"
)
```
